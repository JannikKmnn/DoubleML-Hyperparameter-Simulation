```python
from utils import make_CCDDHNR2018_data_list

n_rep = 1000
n_obs = 500
n_vars = 20
alpha = 0.5

data = make_CCDDHNR2018_data_list(n_rep=n_rep, n_obs=n_obs, n_vars=n_vars, alpha=alpha)
```

```python
print(f"Length of Data List: {len(data)}")
print(f"Number of Arrays for (x, y, d): {len(data[0])}")
print(f"Shape of x: {data[0][0].shape}")
print(f"Shape of y: {data[0][1].shape}")
print(f"Shape of d: {data[0][2].shape}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x_1 = np.array([i[0] for i in data[0][0]])
x_3 = np.array([i[2] for i in data[0][0]])
X_1, X_3 = np.meshgrid(x_1, x_3)
m_0 = X_1 + (1/4)*(np.exp(X_3)/(1+np.exp(X_3)))

fig = plt.figure(figsize=(18,18))
ax = plt.axes(projection='3d')
ax.plot_surface(X_1, X_3, m_0, cmap='RdYlBu')
ax.set_xlabel('X_1')
ax.set_ylabel('X_3')
ax.set_zlabel('m_0')
ax.view_init(elev=30, azim=90)
plt.show()
```
